```python
import pandas as pd
import numpy as np
url = '/2018_Central_Park_Squirrel_Census_-_Squirrel_Data_20250721 (1).csv'
squirrel_df = pd.read_csv(url)
squirrel_df.head()
```

```python
#Exploring data
squirrel_df.info()

'''
columns with MANY  null values:
Color Notes 182 non-null, Specific Location 476 non-null,
Other activities 437 non null, Other Interactions 437 non-null,

columns with few null values:
Age 2902 non-null, Primary Fur Color 2968 non-null, Highlight Fur Color 1937 non-null,
Location 2959 non-null, Above Ground Sighter Measurement 2909 non-null

'''
```

```python
squirrel_df.columns
```

```python
#Dropping columns with exessive null values to clean the data
cleaned_squirrel_df = squirrel_df.drop(columns=['Color notes', 'Specific Location', 'Other Activities', 'Other Interactions'])
cleaned_squirrel_df.info()
```

```python
#EXPLORING COLUMNS WITH MOST DATA NON-NULL
cleaned_squirrel_df['Age'].value_counts()
```

```python
cleaned_squirrel_df['Highlight Fur Color'].value_counts()
```

```python
cleaned_squirrel_df['Primary Fur Color'].value_counts()
```

```python
cleaned_squirrel_df['Above Ground Sighter Measurement'].unique()
```

```python
cleaned_squirrel_df['Location'].value_counts()
```

```python
cleaned_squirrel_df['Age'] = cleaned_squirrel_df['Age'].fillna(value ='Unknown')
cleaned_squirrel_df['Highlight Fur Color'] = cleaned_squirrel_df['Highlight Fur Color'].fillna(value ='Unknown')
cleaned_squirrel_df['Primary Fur Color'] = cleaned_squirrel_df['Primary Fur Color'].fillna(value ='Unknown')
cleaned_squirrel_df['Above Ground Sighter Measurement'] = cleaned_squirrel_df['Above Ground Sighter Measurement'].fillna(value ='Unknown')
cleaned_squirrel_df['Location'] = cleaned_squirrel_df['Location'].fillna(value ='Unknown')
cleaned_squirrel_df
```
